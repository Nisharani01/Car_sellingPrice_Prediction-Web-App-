```python
data.drop('Year',axis=1,inplace=True)
```

```python
# Outlier Removal
```

```python
import seaborn as sns
```

```python
sns.boxplot(data['Selling_Price'])
```

```python
sorted(data['Selling_Price'],reverse=True)
```

```python
data = data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
```

```python
data.shape
```

```python
# Encoding the Categorical Columns
```

```python
data.head(1)
```

```python
data['Fuel_Type'].unique()
```

```python
data['Fuel_Type'] = data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
data['Fuel_Type'].unique()
```

```python
data['Seller_Type'].unique()
```

```python
data['Seller_Type'] = data['Seller_Type'].map({'Dealer':0,'Individual':1})
```

```python
data['Seller_Type'].unique()
```

```python
data['Transmission'].unique()
```

```python
data['Transmission'] =data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
data['Transmission'].unique()
```

```python
data.head()
```

```python
# X,y 
```

```python
X = data.drop(['Car_Name','Selling_Price'],axis=1)
y = data['Selling_Price']
```

```python
# train_test_split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)

RFR=RandomForestRegressor()
RFR.fit(x_train,y_train)

GBR=GradientBoostingRegressor()
GBR.fit(x_train,y_train)

xgb=XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
# predicting test data

pred_1=LR.predict(x_test)
pred_2=RFR.predict(x_test)
pred_3=GBR.predict(x_test)
pred_4=xgb.predict(x_test)
```

```python
# Evaluating the algorithm


from sklearn.metrics import r2_score
```

```python
score_1=r2_score(y_test,pred_1)
score_2=r2_score(y_test,pred_2)
score_3=r2_score(y_test,pred_3)
score_4=r2_score(y_test,pred_4)
```

```python
final_score=pd.DataFrame({'Models':['LR','RFR','GBR','XGB'],'r2_score':[score_1,score_2,score_3,score_4]})
```

```python
print(final_score)
```

```python
sns.barplot(x=final_score['Models'],y=final_score['r2_score'])
```

```python
xgb=XGBRegressor()
xgb_final=xgb.fit(X,y)
```

```python
import joblib
```

```python
joblib.dump(xgb_final,'car_price_pred')
```
